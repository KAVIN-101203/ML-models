```python
import pandas as pd
```

```python
df = pd.read_csv('Melbourne.csv')
```

```python
df
```

```python
col = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
df = df[col]
```

```python
df
```

```python
df.nunique()
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
df = pd.get_dummies(df,dtype = 'int')
x = df.drop(columns = 'Price')
y = df['Price']
```

```python
df
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import Lasso
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30)
```

```python
reg1 = Lasso(alpha=50, max_iter = 100, tol = 0.1)
reg1.fit(x_train,y_train)
```

```python
reg1.score(x_train,y_train)
```

```python
reg1.score(x_test,y_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x1_train,x1_test,y1_train,y1_test = train_test_split(x,y,test_size =0.30)
model = LinearRegression()
model.fit(x1_train,y1_train)
```

```python
model.score(x1_train,y1_train)
```

```python
model.score(x1_test,y1_test)
```
