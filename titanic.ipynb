```python
a = int(input("Age: "))
b = int(input("Parch: "))
c = float(input("fare: "))
d = input("Gender: ")
j = input("embarked: ")
e=0
f=0
g=0
h=0
i=0
if(d == "MALE" or "male"):
    e,f = 0,1
elif(d == "FEMALE" or "female"):
    e,f = 1,0
if(j == "C"):
    g,h,i=1,0,0
if(j == "Q"):
    g,h,i=0,1,0
if(j == "S"):
    g,h,i=0,0,1
k = model.predict([[a,b,c,e,f,g,h,i]])[0]
if(k == 0):
    print("died")
if(k == 1):
    print("Survied")
```
